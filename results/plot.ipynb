```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('timings8.csv')
df['size'] = df['file'].apply(lambda x: int(x.split('_')[-1].split('.')[0]))
df.sort_values('size', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['size'], df['improved__grep_avg_time_us'], label='Improved Avg Time (ms)', marker='o')
plt.plot(df['size'], df['original_grep_avg_time_us'], label='Original Avg Time (ms)', marker='x')

plt.title('Performance Analysis: REGEX b?^na^n before and after Suffix Extraction')
plt.xlabel('Size of REGEX Pattern')
plt.ylabel('Average Execution Time (ms)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
